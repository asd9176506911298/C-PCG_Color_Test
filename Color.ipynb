```python
import json
from opencc import OpenCC
from skimage import io

cc = OpenCC('s2twp')

def translate(str):
    return (cc.convert(str))

with open('data.json', mode='r', encoding='utf-8') as file:
    data=json.load(file)
print(data.keys())
for i in data.keys():
    d = data[i]
    print(translate(i))
    print(translate('{}%的人特別喜歡{}色'.format(d['percent'],i)))
    print(translate('每{}個人中就會有一個喜歡你'.format(d['peoplecount'])))
    print(translate('你的特質是:{}'.format(d['oneWord'])))
    print(translate('在別人眼中你是{}色:\n{}'.format(d['otherPeople'],d['otherPeopleText'])))
    print()
    print(translate('本質上你是{}色:\n{}'.format(d['infact'], d['infactText'])))
    print()
    print('這些顏色被你吸引')
    for x in d['attracted']:
        print(translate(x), end=' ')
    print('\n')
    print('這些顏色請保持距離')
    for x in d['keepAway']:
        print(translate(x), end=' ')
    print()
    print('songName:',d['songName'])
    print('songId:',d['songId'])
    print('songUrl:',d['songUrl'])
    print('keywordOther:',translate(d['keywordOther']))
    print('keywordInfact:',translate(d['keywordInfact']))
    print('sharePic:',d['sharePic'])
    image = io.imread(d['sharePic'])
    io.imshow(image)
    io.show()
```
